```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('pexels-pixabay-56866.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Get image dimensions
height, width = image.shape[:2]
```

```python
# Scaling
scaled_image = cv2.resize(image, (int(width * 1.5), int(height * 1.5)))
```

```python
angle = 45
rotation_matrix = cv2.getRotationMatrix2D((width//2, height//2), angle, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python

# Translation (Shifting)
tx, ty = 50, 100
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
shifted_image = cv2.warpAffine(image, translation_matrix, (width, height))
```

```python
# Reflection (Mirroring)
reflected_image = cv2.flip(image, 1)  # Horizontal flip
```

```python

# Shearing
shear_matrix = np.float32([[1, 0.5, 0], [0.5, 1, 0]])
sheared_image = cv2.warpAffine(image, shear_matrix, (width, height))
```

```python
# Affine Transformation
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])
affine_matrix = cv2.getAffineTransform(pts1, pts2)
affine_transformed = cv2.warpAffine(image, affine_matrix, (width, height))
```

```python
# Function to display images
def show_images(images, titles):
    plt.figure(figsize=(12, 8))
    for i in range(len(images)):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])
        plt.title(titles[i])
        plt.axis('off')
    plt.show()
```

```python
# Display all transformations
show_images(
    [image, scaled_image, rotated_image, shifted_image, reflected_image, sheared_image, affine_transformed],
    ["Original", "Scaled", "Rotated", "Shifted", "Reflected", "Sheared", "Affine Transformed"]
)
```
