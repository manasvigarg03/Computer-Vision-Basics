```python

import cv2
import numpy as np
from matplotlib import pyplot as plt
def perspective_transform(f, x1, y1, x2, y2, x3, y3, x4, y4, width, height):
    img = cv2.imread(f)
    if img is None:
        raise ValueError("Error")
    src_pts = np.array([[x1, y1], [x2, y2], [x3, y3], [x4, y4]], dtype=np.float32)
    dst_pts = np.array([[0, 0], 
                        [width - 1, 0], 
                        [0, height - 1], 
                        [width - 1, height - 1]], dtype=np.float32)
    M = cv2.getPerspectiveTransform(src_pts, dst_pts)
    transformed_image = cv2.warpPerspective(img, M, (width, height))
    return transformed_image

transformed_img = perspective_transform("pexels-pixabay-56866.jpg", 100, 200, 400, 200, 100, 600, 400, 600, 300, 400)
image = cv2.imread("pexels-pixabay-56866.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
transformed_img = cv2.cvtColor(transformed_img, cv2.COLOR_BGR2RGB)
plt.subplot(121)
plt.imshow(image)
plt.title('Input')
plt.subplot(122)
plt.imshow(transformed_img)
plt.title('Output')
plt.show()
#cv2.imwrite("output.jpg", transformed_img)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def perspective_transform(img, pt1, pt2):
    rows, cols, ch = img.shape
    # Calculate the perspective transformation matrix
    matrix_aff = cv2.getPerspectiveTransform(pt1, pt2)
    
    # Apply the perspective transformation
    dst = cv2.warpPerspective(img, matrix_aff, (cols, rows))
    
    # Display the input and output images
    plt.subplot(121), plt.imshow(img), plt.title('Input')
    plt.subplot(122), plt.imshow(dst), plt.title('Output')
    plt.show()

# Load the image
img = cv2.imread('61jHzpZzEYL._AC_SL1498_.jpg')

# Define the points on the input image
pt1 = np.float32([[50, 65], [370, 52], [30, 387], [390, 390]])

# Define the points on the output image (rectangular area)
pt2 = np.float32([[0, 0], [310, 0], [0, 310], [310, 310]])

# Apply perspective transformation
perspective_transform(img, pt1, pt2)
```
